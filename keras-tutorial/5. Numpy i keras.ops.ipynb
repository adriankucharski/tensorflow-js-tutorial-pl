## `keras.ops`

Keras oferuje moduł `keras.ops`, który zawiera operacje w większości zgodne z operacjami z biblioteki `numpy`. Różnica jest taka, że operacje z `keras.ops` działają na tensorach i pozwalają na propagację wsteczną.

```python
from keras import ops
import matplotlib.pyplot as plt

# Tensor
a = ops.ones((32, 32, 3))
plt.matshow(a)
plt.show()

# Macierz diagonalna
b = ops.eye(5)
plt.matshow(b)
plt.show()

# Argmax i argmin
b = ops.eye(5)
c = ops.argmax(b, axis=1)
print("c.shape", c.shape, c)

# I wiele wiele innych operacji dokładnie takich samych jak w numpy
```

### `keras.ops.image` - działanie na obrazach

```python
from keras import ops
import matplotlib.pyplot as plt

# Zmiana kształtu
b = ops.eye(5)
print("b.shape", b.shape)
b = ops.expand_dims(b, axis=-1)
print("b.shape + expand_dims", b.shape)
d = ops.image.resize(b, (64, 64))
plt.matshow(d)
plt.show()

b = ops.eye(5)[None, ..., None]
images = ops.image.resize(b, (64, 64))
patches = ops.image.extract_patches(images, (16, 16), padding="valid")

print("images.shape", images.shape)
print("patches.shape", patches.shape)

patch = patches[0, 0, 0]
plt.matshow(ops.reshape(patch, (16, 16)))
plt.show()
```

### `keras.ops.linalg` - operacje na macierzach

```python
from keras import ops
import matplotlib.pyplot as plt

# Macierz diagonalna
a = ops.eye(16)
b = ops.ones((16, 16)) * 0.5

# Dodawanie
c = ops.add(a, b)  # lub a + b
# Mnożenie
d = ops.multiply(a, b)  # lub a * b
# Dzielenie
e = ops.divide(a, b)  # lub a / b
# Potęgowanie
f = ops.power(a, 2)  # lub a ** 2


# Mnożenie macierzowe
g = ops.dot(a, b)  # lub a @ b
# Wartości własne i wektory własne
h = ops.linalg.eig(ops.expand_dims(g, axis=0))

# Odwracanie macierzy
i = ops.linalg.inv(ops.eye(5))
```
